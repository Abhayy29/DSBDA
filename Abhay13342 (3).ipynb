```python
import numpy as np
Q1 = np.percentile(df['Math_Score'], 25)
Q3 = np.percentile(df['Math_Score'], 75)

IQR = Q3 - Q1
```

```python
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

print("Lower Limit:", lower)
print("Upper Limit:", upper)
```

```python
outliers = df[(df['Math_Score'] < upper) | (df['Math_Score'] > lower)]
print(outliers)
```

```python
df = df.drop(df[(df['Math_Score'] < lower) | 
                (df['Math_Score'] > upper)].index)
```

```python
import matplotlib.pyplot as plt

df.boxplot(column=['Math_Score',
                   'Reading_Score',
                   'Writing_Score',
                   'Placement_Score'])

plt.show()
```
