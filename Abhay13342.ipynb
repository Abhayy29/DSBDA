```python
import numpy as np

b = np.array([(1.5, 2, 3), (4, 5, 6)], dtype=float)

print("dtype:", b.dtype)
print("dtype name:", b.dtype.name)

b_int = b.astype(int)
print("\nConverted to int:")
print(b_int)
```

```python
a = np.array([1, 2, 3])
b = np.array([(1.5, 2, 3), (4, 5, 6)], dtype=float)

add_result = a + b
print("a + b =\n", add_result)

sub_result = a - b
print("\na - b =\n", sub_result)
```

```python
b = np.array([(1.5, 2, 3), (4, 5, 6)], dtype=float)
mul_result = b * b
div_result = b / b

print("b * b =\n", mul_result)
print("\nb / b =\n", div_result)
```

```python
import numpy as np

b = np.array([(1.5, 2, 3), (4, 5, 6)], dtype=float)

print("Sum:", np.sum(b))
print("Min:", np.min(b))
print("Max:", np.max(b))
print("Cumulative Sum:\n", np.cumsum(b))
print("Mean:", np.mean(b))
print("Median:", np.median(b))
print("Standard Deviation:", np.std(b))
```
